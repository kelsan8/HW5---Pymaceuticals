```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy as sp

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_data = "data/mouse_drug_data.csv"
clinical_data = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_data)
clinical_df = pd.read_csv(clinical_data)

# Combine the data into a single dataset
merge_table = pd.merge(mouse_df, clinical_df, on="Mouse ID")


# Display the data table for preview
merge_table.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
# Group by the relevant groups then calculate the means for the timepoint and drug
drug_timing_df = merge_table.groupby(["Drug", "Timepoint"])
tumor_df = drug_timing_df["Tumor Volume (mm3)"].mean()

#drug_timing_df.head() --- checking the initial outcome of the grouping
# Ensure the means are arranged into a DataFrame
tumor_df = pd.DataFrame(tumor_df)

# Preview DataFrame
tumor_df.head(10)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Uses the previous DataFrame made with the groupby function
tumor_df = drug_timing_df["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
tumor_sem_df = pd.DataFrame(tumor_df)

# Preview DataFrame
tumor_sem_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Sets the columns to the individual drug names to compare their averages at each timepoint individually
tumor_sem_df.columns = tumor_sem_df.iloc[0]

# Preview that Reformatting worked
tumor_df = pd.DataFrame(tumor_sem_df)
```

```python
# Generate the Plot (with Error Bars)

# Set the plot and figure size of the plot
plt.figure(figsize=(10,9))

# Set the values for the x-axis to match the timepoints in the data set
time_x_axis = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]

# Create separate data frames for each drug and generate the plot points for mean and sem of the tumor volume
# Not all of the drugs are used in the plot, but the data is generated if needed in later plots
capomulin_info = merge_table.loc[merge_table["Drug"]=="Capomulin", :]
capomulin = capomulin_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
capomulin_sem = capomulin_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()
capomulin_plot = plt.errorbar(time_x_axis, capomulin, yerr=capomulin_sem, fmt="o", ls="dashed", color="red", label="Capomulin")

ceftamin_info = merge_table.loc[merge_table["Drug"]=="Ceftamin", :]
ceftamin = ceftamin_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
ceftamin_sem = ceftamin_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

infubinol_info = merge_table.loc[merge_table["Drug"]=="Infubinol", :]
infubinol = infubinol_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
infubinol_sem = infubinol_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()
infubinol_plot = plt.errorbar(time_x_axis, infubinol, yerr=infubinol_sem, fmt="^", ls="dashed", color="blue", label="Infubinol")

ketapril_info = merge_table.loc[merge_table["Drug"]=="Ketapril", :]
ketapril = ketapril_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
ketapril_sem = ketapril_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()
ketapril_plot = plt.errorbar(time_x_axis, ketapril, yerr=ketapril_sem, fmt="s", ls="dashed", color="green", label="Ketapril")

naftisol_info = merge_table.loc[merge_table["Drug"]=="Naftisol", :]
naftisol = naftisol_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
naftisol_sem = naftisol_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

placebo_info = merge_table.loc[merge_table["Drug"]=="Placebo", :]
placebo = placebo_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
placebo_sem = placebo_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()
placebo_plot = plt.errorbar(time_x_axis, placebo, yerr=placebo_sem, fmt="D", ls="dashed", color="black", label="Placebo")

propriva_info = merge_table.loc[merge_table["Drug"]=="Propriva", :]
propriva = propriva_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
propriva_sem = propriva_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

ramicane_info = merge_table.loc[merge_table["Drug"]=="Ramicane", :]
ramicane = ramicane_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
ramicane_sem = ramicane_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

stelasyn_info = merge_table.loc[merge_table["Drug"]=="Stelasyn", :]
stelasyn = stelasyn_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
stelasyn_sem = stelasyn_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

zoniferol_info = merge_table.loc[merge_table["Drug"]=="Zoniferol", :]
zoniferol = zoniferol_info.groupby("Timepoint")["Tumor Volume (mm3)"].mean()
zoniferol_sem = zoniferol_info.groupby("Timepoint")["Tumor Volume (mm3)"].sem()

# Add a title to the graph
plt.title("Tumor Response to Treatment")
# Add a label for the x-axis
plt.xlabel("Time Duration of Treatment (Days)")
# Add a label for the y-axis
plt.ylabel("Average Tumor Volume (mm3) Over Time")
# Add a legend to the graph in the top left
plt.legend(loc='upper left')
# Add a grid to the graph
plt.grid()

# Save the Figure
plt.savefig("../Images/mean_tumor_volume_over_time")

plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice = merge_table.groupby(["Drug", "Timepoint"])
mouse_count = mice.count()['Mouse ID']
# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_count)

# Preview DataFrame
mouse_count_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_df.columns = mouse_count_df.iloc[0]
mouse_count_df = mouse_count_df.groupby("Timepoint")
# Preview the Data Frame
mouse_count_df
```

```python
# Generate the Plot (Accounting for percentages)

# Set the plot and figure size of the plot
plt.figure(figsize=(10,9))

# Set the values for the x-axis to match the timepoints in the data set
time_x_axis = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]

# Calculate the percentage survival rate for each drug and set up the plots
capomulin_info = merge_table.loc[merge_table["Drug"]=="Capomulin", :]
capomulin = capomulin_info.groupby("Timepoint").count()["Mouse ID"]
cap_percent = (capomulin / 25) * 100
capomulin_plot = plt.errorbar(time_x_axis, cap_percent, fmt="o", ls="dashed", color="red", label="Capomulin")

infubinol_info = merge_table.loc[merge_table["Drug"]=="Infubinol", :]
infubinol = infubinol_info.groupby("Timepoint").count()["Mouse ID"]
inf_percent = (infubinol / 25) * 100
infubinol_plot = plt.errorbar(time_x_axis, inf_percent, fmt="^", ls="dashed", color="blue", label="Infubinol")

ketapril_info = merge_table.loc[merge_table["Drug"]=="Ketapril", :]
ketapril = ketapril_info.groupby("Timepoint").count()["Mouse ID"]
ket_percent = (ketapril / 25) * 100
ketapril_plot = plt.errorbar(time_x_axis, ket_percent, fmt="s", ls="dashed", color="green", label="Ketapril")

placebo_info = merge_table.loc[merge_table["Drug"]=="Placebo", :]
placebo = placebo_info.groupby("Timepoint").count()["Mouse ID"]
plc_percent = (placebo / 25) * 100
placebo_plot = plt.errorbar(time_x_axis, plc_percent, fmt="D", ls="dashed", color="black", label="Placebo")

# Add a title to the graph
plt.title("Percent Survival Over Time")
# Add a label for the x-axis
plt.xlabel("Time Duration of Treatment (Days)")
# Add a label for the y-axis
plt.ylabel("Survival Rate (%)")
# Add a legend to the graph in the top left
plt.legend(loc='upper left')
# Add a grid to the graph
plt.grid()

# Save the Figure
plt.savefig("../Images/percent_survival_over_time")

# Show the Figure
plt.show()
```
